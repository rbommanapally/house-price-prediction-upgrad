```python
## lets Remove any columns of object type with more than 80% unique values

def removeUniqueValuesForObjectType(data, prcnt):
    df1= data.copy()
    my_list= []
    cat_obj_cols = df1.select_dtypes(include=['object'])
    for col in (cat_obj_cols):
        if(df1[col].value_counts().max() > int(prcnt*len(df1.index)/100)):
            my_list.append(col)
    return my_list

##Removing the skewed data
objColumnsToDrop = removeUniqueValuesForObjectType(data=df, prcnt=80)  
print("Categorical object type columns removed:",objColumnsToDrop)
print("\nNumber of categorical object type columns removed: ", len(objColumnsToDrop),"\n")
```

```python
#lets drop all the above columns which more than 80 % unique values, as this might not help in preditcing a good model
print("shape Before dropping ", df.shape)
df = df.drop(objColumnsToDrop, axis=1)
print("shape After dropping ", df.shape)
```

```python
## lets Remove any columns of numerical type with more than 80% unique values

def removeUniqueValuesForNumericType(data, prcnt):
    df1= data.copy()
    my_list= []
    num_cols = df1.select_dtypes(include=['int64','float64','int32','float32'])
    for col in (num_cols):
        if(df1[col].value_counts().max() > int(prcnt*len(df1.index)/100)):
            my_list.append(col)
    return my_list

##Removing the skewed data
numColumnsToDrop = removeUniqueValuesForNumericType(data=df, prcnt=80)  
print("Numeric columns removed:",numColumnsToDrop)
print("\nNumber of numeric columns removed: ", len(numColumnsToDrop),"\n")

```

```python
#lets drop the numeric columns with more than 80% unqiue values
print("shape Before dropping ", df.shape)
df = df.drop(numColumnsToDrop, axis=1)
print("shape After dropping ", df.shape)
```

## Skewed Data Handling

```python
# lets see if the numerical columns data has any skewed data
df[numcols]
```

```python
df[numcols].skew()
```

```python
# lets handled skew data
num_cols_skewed = df[numcols].skew()
#we need to rectrict data between 0 and 1, lets handle the other data
num_cols_skewed = num_cols_skewed[(num_cols_skewed>=1)|(num_cols_skewed<=-1)]
```

```python
num_cols_skewed
```

```python
# Get skewed columns
skewedcols = num_cols_skewed.index.tolist()
skewedcols
```

```python
# Lets perform log correction to skewed columns

chk = abs(np.log(df[skewedcols]).skew())
logcols = chk[(chk<=1)&(chk>=-1)].index.tolist()
logcols
```

```python
# lets check the remaining cols after correction
balcolms = [x for x in skewedcols if x not in logcols]
balcolms
```

```python
# lets perfome square root correction to balance columns
chk = abs(np.sqrt(df[balcolms]).skew())
balsqrtcolms = chk[(chk<=1) & (chk>=-1)].index.tolist()
balsqrtcolms
```

```python
# lets perform on comeplete df
df.skew()
```

```python
# Transform columns with high skew

df_orig = df.copy()
df[logcols] = np.log(df[logcols])
df[balsqrtcolms] = np.sqrt(df[balsqrtcolms])
df.skew()
```
